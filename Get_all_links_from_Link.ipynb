```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
d =[]
links = []
li=[]
response = requests.get("https://www.youm7.com/")
soup = BeautifulSoup(response.content, 'lxml')

```

```python
z=[]
x= soup.find_all('a')
for i in range(len(x)):
    if x[i].attrs['href'] not in z:        
        z.append(x[i].attrs['href'])
    else:
        continue
len(z)
```

```python
z
```

```python
df_dict={'youm7': z}
df = pd.DataFrame(df_dict)
df.head(30)
```

```python
df.to_csv('youm7_0.csv', encoding='utf-8' ,index=False)
```
